```python
%matplotlib notebook
import argparse
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/Users/alexzl/tmp/ER/data/CrimesAggregated.csv", delimiter=",")
data["District"] = data["District"].astype(int)
data[:20]
```

```python
df=data.pivot(index="Primary Type", columns="District", values="0").fillna(0)
df
```

```python
colormap = plt.cm.cubehelix_r
sns.set(font_scale = 0.7)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
ax.xaxis.set_label_position('top') 

ax.invert_yaxis()
ax.xaxis.tick_top()
h=sns.heatmap(ax=ax, data=df, cmap=colormap,xticklabels=True,yticklabels=True)
h.xaxis.tick_top()
h.set_xticklabels(h.get_xticklabels(), rotation=90)
fig.subplots_adjust(left=0.2)
```
